```python
import pandas as pd
# https://www.kaggle.com/code/toastymcvoid/analyzing-video-game-sales
df=pd.read_csv('vgsales.csv')
df.head(9)

```

```python
X = df.select_dtypes(include=[object])
X.head(3)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X_2 = X.apply(le.fit_transform)
X_2 = X_2.drop(columns=['Name'])
X_2.head(3)
```

```python
y = df.drop(columns=['Rank','Name','Genre','Year','Publisher','Platform'])
y
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X_2,y,test_size=0.25)
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
predictions=model.predict(X_test)
score = model.score(X_test,y_test)
score
```

```python
import joblib

joblib.dump(model,'output-model.joblib')
```
